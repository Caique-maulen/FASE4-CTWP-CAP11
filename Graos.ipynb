```python
import pandas as pd

# Caminho para o arquivo
caminho_arquivo = r'C:\Users\PC\Downloads\seeds\seeds_dataset.txt'

# Lendo e processando os dados
with open(caminho_arquivo, 'r') as arquivo:
    linhas = arquivo.readlines()

# Convertendo para float
dados = [[float(valor) for valor in linha.strip().split()] for linha in linhas]

# Criando o DataFrame
df = pd.DataFrame(dados)

# Renomeando as colunas
df.columns = [
    "Área",
    "Perímetro",
    "Compacidade",
    "Comprimento do kernel",
    "Largura do kernel",
    "Assimetria",
    "Comprimento da ranhura",
    "Classe"
]

# Calculando estatísticas
for coluna in df.columns[:-1]:  # Ignora "Classe" se for categórica
    media = df[coluna].mean()
    mediana = df[coluna].median()
    desvio = df[coluna].std()

    print(f"{coluna}:")
    print(f"  Média: {media:.2f}")
    print(f"  Mediana: {mediana:.2f}")
    print(f"  Desvio padrão: {desvio:.2f}")
    print("-" * 30)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# === 1. Ler e preparar os dados ===
caminho_arquivo = r'C:\Users\PC\Downloads\seeds\seeds_dataset.txt'

with open(caminho_arquivo, 'r') as arquivo:
    linhas = arquivo.readlines()

dados = [[float(valor) for valor in linha.strip().split()] for linha in linhas]
df = pd.DataFrame(dados)

# Renomear colunas
df.columns = [
    "Área",
    "Perímetro",
    "Compacidade",
    "Comprimento do kernel",
    "Largura do kernel",
    "Assimetria",
    "Comprimento da ranhura",
    "Classe"
]
df["Classe"] = df["Classe"].astype(int)  # Garantir que a classe seja inteira

# === 2. Gráficos de dispersão entre características ===
sns.pairplot(df, hue="Classe", palette="Set2", diag_kind="kde")
plt.suptitle("Gráficos de Dispersão entre Características", y=1.02)
plt.show()

# === 3. Verificar e tratar valores ausentes ===
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Se houver valores ausentes, preencher com a média (ou usar dropna())
if df.isnull().values.any():
    df.fillna(df.mean(), inplace=True)
    print("Valores ausentes preenchidos com a média.")

# === 4. Padronizar as características ===
X = df.drop("Classe", axis=1)
y = df["Classe"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Converter para novo DataFrame com as mesmas colunas
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)
df_scaled["Classe"] = y.values

print("\nPrimeiras linhas dos dados padronizados:")
print(df_scaled.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt

# === 1. Carregar e preparar os dados (caso ainda não esteja feito) ===
caminho_arquivo = r'C:\Users\PC\Downloads\seeds\seeds_dataset.txt'
with open(caminho_arquivo, 'r') as arquivo:
    linhas = arquivo.readlines()

dados = [[float(valor) for valor in linha.strip().split()] for linha in linhas]
df = pd.DataFrame(dados)
df.columns = [
    "Área", "Perímetro", "Compacidade", "Comprimento do kernel",
    "Largura do kernel", "Assimetria", "Comprimento da ranhura", "Classe"
]
df["Classe"] = df["Classe"].astype(int)

# === 2. Separar X e y, padronizar os dados ===
X = df.drop("Classe", axis=1)
y = df["Classe"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === 3. Dividir em treino (70%) e teste (30%) ===
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# === 4. Definir os modelos ===
modelos = {
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42)
}

# === 5. Treinar, avaliar e comparar ===
for nome, modelo in modelos.items():
    print(f"\n=== {nome} ===")
    modelo.fit(X_train, y_train)
    y_pred = modelo.predict(X_test)

    # Relatório de métricas
    print(classification_report(y_test, y_pred, digits=3))

    # Matriz de confusão
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Matriz de Confusão - {nome}")
    plt.xlabel("Predito")
    plt.ylabel("Real")
    plt.tight_layout()
    plt.show()
```
